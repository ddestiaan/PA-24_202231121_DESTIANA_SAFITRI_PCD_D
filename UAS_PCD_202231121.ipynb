```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Baca gambar
img = cv2.imread('desti.jpg')

# menampilkanm gambar dalam bentuk jendela window 
cv2.imshow("Menampilkan Gambar Tepi", img) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
# Konversi gambar ke grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Terapkan Gaussian Blur untuk mengurangi noise
blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)

#MEMBUAT VARIABLE ADGES UNTUK MENAMPUNG HASIL CANNY(DETEKSI TEPI) DARI VARIABLE IMG
edges =  cv2.Canny (img, 250, 300) #aras keabuan yang diubah antara 250-300

# Deteksi kontur
contours, hierarchy = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Gambar kontur pada gambar asli
contour_image = image.copy()
cv2.drawContours(contour_image, contours, -50, (0, 255, 0), 5)
```

```python
# Output Pertama: Gambar Asli dan Hasil Canny Edge Detection
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Gambar Asli')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Canny Edge Detection')
plt.imshow(edges, cmap='gray')

plt.show()
```

```python
# Output Kedua: Gambar Asli, Hasil Canny Edge Detection, dan Hasil Countours Detection
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Gambar Asli')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title('Canny Edge Detection')
plt.imshow(edges, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Countours Detection')
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))

plt.show()
```
